```python
total_sales=df.groupby(['Product ID','Category'])['Sales'].sum() #grouped 2 columns based on the third column and performed summation
avg_sales=df.groupby(['Product ID','Category'])['Sales'].mean()  #grouped 2 columns based on the third column and performed mean
```

```python
print(total_sales) #printing total sales
```

```python
plots=df.groupby(['Product ID','Category'])['Sales'].sum().reset_index() #ploting graph by grouping 2 columns based on sum of sales .
```

```python
print(plots)
```

```python
import matplotlib.pyplot as  plt #imported matplotlib for visualization
```

```python
type_sales = df.groupby('Category')['Sales'].mean().reset_index()
print(type_sales)
type_sales.plot(kind='bar', x='Category', y='Sales', figsize=(6,4))
plt.xlabel('Category')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

Observations:
We can observe that Technology is having highest sales followed by least sales in Office Supplies.

```python
import matplotlib.pyplot as plt
```

```python
plt_data= df.groupby('Quantity')['Sales'].sum().reset_index()
print('pie_data')
plt_data.plot(kind='bar', x='Quantity', y='Sales', figsize=(10,4))
plt.xlabel('Quantity')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

Observations:
We can observe that Quantity 3 is having highest sales and quantity 12 is having the least sales .
