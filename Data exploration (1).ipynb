```python
import matplotlib.pyplot as plt
```

```python
type_sales = df.groupby('Category')['Sales'].mean().reset_index()
print(type_sales)
type_sales.plot(kind='bar', x='Category', y='Sales', figsize=(6,4))
plt.xlabel('Category')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

```python
QUANTITY SOLD BASED ON SALES
```

```python
import matplotlib.pyplot as plt
```

```python
pie_data= df.groupby('Quantity')['Sales'].sum().reset_index()
print('pie_data')
pie_data.plot(kind='bar', x='Quantity', y='Sales', figsize=(10,4))
plt.xlabel('Quantity')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```
