```python
total_sales=df.groupby(['Item_Type','Outlet_Type', 'Outlet_Location_Type'])['Item_Outlet_Sales'].sum()
avg_sales=df.groupby(['Product ID','Category'])['Sales'].mean()
print(total_sales)
```

```python
plots=df.groupby(['Product ID','Category'])['Sales'].sum().reset_index()
print(plots)
```

```python
import matplotlib.pyplot as plt
```

```python
type_sales = df.groupby('Category')['Sales'].mean().reset_index()
print(type_sales)
type_sales.plot(kind='bar', x='Category', y='Sales', figsize=(6,4))
plt.xlabel('Category')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
pie_data= df.groupby('Quantity')['Sales'].sum().reset_index()
print('pie_data')
pie_data.plot(kind='bar', x='Quantity', y='Sales', figsize=(10,4))
plt.xlabel('Quantity')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```
