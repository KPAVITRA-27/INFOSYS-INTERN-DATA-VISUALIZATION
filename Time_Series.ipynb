ARIMA MODEL:
Used to forecast time series for stationary data.
Denoted using 3 parameters:(p,d,q)
p:no of time lags
d:no of times the data have had past values subtracted.
q:the order of moving average model.
*autocorrelation
*short term forecast

```python
from statsmodels.tsa.arima.model import ARIMA #importing ARIMA
from sklearn.metrics import mean_squared_error
```

STEPS TO PERFORM:
1.Make index as Datetime
2.In order to prepare the ARIMA model we use the sales column.
3.Splitting the data into training data and test data.(eg.80% train 20%test)
4.Determining the p,d,q values
5.Making predictions
6.Evaluating the model use rms
7.Plotting

```python
df.set_index(df['Order Date'])
sales_data=df['Sales']
```

```python
train_data = sales_data[:int(len(sales_data) * 0.8)]
test_data = sales_data[int(len(sales_data) * 0.8):]
```

assuming p:5
         d:1
         q:0

```python
model = ARIMA(train_data, order=(5, 1, 0))
model_fit = model.fit()
```

```python
predictions = model_fit.predict(start=len(train_data), end=len(sales_data)-1)
rmse = np.sqrt(mean_squared_error(test_data, predictions))
print(f"RMSE: {rmse}")
```

```python
plt.plot(test_data, label='Actual')
plt.plot(predictions, label='Predictions')
plt.legend()
plt.show()
```
